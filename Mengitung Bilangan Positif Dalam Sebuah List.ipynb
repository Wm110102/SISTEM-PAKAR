```python
def hitung_bilangan_positif(data):
    if not isinstance(data, list):
        raise ValueError("Input harus berupa list")

    jumlah_positif = 0
    bilangan_positif = []  # Menyimpan bilangan positif dalam list

    for elemen in data:
        if isinstance(elemen, int) and elemen > 0:
            jumlah_positif += 1
            bilangan_positif.append(elemen)

    return jumlah_positif, bilangan_positif


print("========================================")
print("PROGRAM HITUNG BILANGAN POSITIF DALAM LIST")
print("========================================")
print("Nama     : Wahyudi Malihi")
print("NIM      : E1E120099")
print("----------------------------------------")

while True:
    input_data = input("Masukkan elemen-elemen list (pisahkan dengan koma): ")
    data_list = [int(elem) for elem in input_data.split(",")]

    try:
        hasil_jumlah, bilangan_positif = hitung_bilangan_positif(data_list)
        if hasil_jumlah > 0:
            print("\n=== Hasil Perhitungan ===")
            print("+------------------------+")
            print("|         List           |")
            print("+------------------------+")
            for elem in data_list:
                print(f"|{elem:>24}|")
            print("+------------------------+")
            print(f"Jumlah Bilangan Positif: {hasil_jumlah:2d}")
            print("Bilangan Positif dalam List: ", end="")
            print(*bilangan_positif, sep=", ")
            print()
        else:
            print("Tidak ada bilangan positif dalam list ini.")
    except ValueError as e:
        print("Error:", str(e))

    ulang = input("Apakah Anda ingin memasukkan inputan ulang? (Ya/Tidak): ")
    if ulang.lower() != 'ya':
        break

print("----------------------------------------")
print("Terimakasih sudah menggunakan program ini.")
print("By: Wahyudi Malihi - E1E120099")
```
